```python
import pandas as pd

# Assuming you have a DataFrame named df
# Read the CSV file into a DataFrame
df = pd.read_csv('/content/WorldCupMatches.csv', index_col=0)
# Step 1: Create a new dataset that only contains games in the year 1986
world_cup_1986 =df==1986

# Step 4: Get the total number of participating countries
participating_countries = len(world_cup_1986)

# Display the result
print(f"The number of countrie's that participated in the 1986 World Cup is/are: {participating_countries}")
```

```python
# Create a new column "Total Goals" by adding the goals from the "Home Team Goals" and "Away Team Goals" columns
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']

# Count the number of matches with 5 goals or more in total
matches_with_5_goals_or_more = df[df['Total Goals'] >= 5]

# Get the total number of such matches
total_matches_with_5_goals_or_more = len(matches_with_5_goals_or_more)

print(f"The number of matches in World Cup history with 5 goals or more is: {total_matches_with_5_goals_or_more}")
```

```python
# Replace "Home Team Name" and "Away Team Name" entries containing "Korea" with "Korea"
df['Home Team Name'] = df['Home Team Name'].replace({'.*Korea.*': 'Korea'}, regex=True)
df['Away Team Name'] = df['Away Team Name'].replace({'.*Korea.*': 'Korea'}, regex=True)

# Replace "Home Team Initials" and "Away Team Initials" entries with "NSK" for "KOR" and "PRK"
df['Home Team Initials'] = df['Home Team Initials'].replace({'KOR': 'NSK', 'PRK': 'NSK'})
df['Away Team Initials'] = df['Away Team Initials'].replace({'KOR': 'NSK', 'PRK': 'NSK'})

# Display the modified DataFrame
df
```
