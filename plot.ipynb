```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
from sklearn import datasets
sns.set()
iris=datasets.load_iris()

df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['Target']=iris.target
df.head()
```

```python
df["Target"]=df["Target"].map(lambda x: "iris Setosa" if x==0 else("iris Versicolor" if(x==1) else "iris virginica"))
```

```python
df['Target'].value_counts
df.head()
```

```python
class_col=df.groupby('Target')["Target"].count()
class_col
```

```python
fig,ax=plt.subplots(figsize=(8,6))
ax.pie(class_col.values,explode=(0,0,0.1),labels=class_col.index,autopct="%1.2f%%",shadow=True)
ax.set_title("IRIS FLOWERS",fontsize=16,fontdict={'color':'olive'})
plt.show()
fig.savefig("irisflowers.jpg")
```

```python
fig,ax=plt.subplots(figsize=(8,6))
sns.barplot(x=class_col.index,y=class_col.values,width=0.4)
for i,j in enumerate(class_col):
  ax.text(i,j+1,str(j),ha=('center'))

plt.title=('Target')
```
